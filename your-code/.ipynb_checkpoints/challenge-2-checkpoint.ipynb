## Coding Session I - 30 minutes

From the Fitbit data, we want to visually understand: 

1. **How the average number of steps change by month. Use the appropriate visualization to show the median steps by month?**
1. **Is Fitbitter more active on weekend or workdays?**

##### All plots must be in the same jupyter notebook cell.

Hints:

- Use `Months_encoded` and `Week_or Weekend` columns
- Use `matplolib.pyplot` object oriented API
- Set your size figure to `12,4`
- Explore `plt.sca`
- Explore `plt.xticks`
- Save your figures

```python
# Driver 1 enter codes here
month_mean = fitbit.groupby('Months_encoded').mean()['Steps']

fig,graphs = plt.subplots(figsize = (10,4),nrows = 1, ncols = 2)
fig.tight_layout(pad=3.0) 
# la lÃ­nea anterior es para dar un espaciado entre las grÃ¡ficas
plt.sca(graphs[0])
plt.style.use('classic')
#print(month_mean,'\n')
month_mean.plot(kind='line');
plt.grid(True)
plt.ylabel('Promedio de pasos p/ mes')
# no supe como hacer un sorted a la columna de 'Months'
# ademÃ¡s de que se observa uqe no empieza justo en el eje la grÃ¡fica
# asÃ­ que se deja un espacio en blanco para que coincidan valores
plt.xticks(list(range(13)),['','Ene','Feb','Mar','Abr','May','Jun','Jul','Ago','Sep','Oct','Nov','Dic'],rotation=45);
plt.sca(graphs[1])
#---------------------------------------------------------------------------
# 0 is a weekend & 1 is a weekday
wknd_wds = fitbit.groupby('Work_or_Weekend').mean()['Steps']
wknd_wds.plot(kind = 'bar')
plt.ylabel('Promedio de pasos')
plt.grid(True)
plt.xticks(list(range(2)),['Fin de semana','Entresemana'], rotation=0);
# notamos que el fin de semana tiene menor cantidad de pasos a comparaciÃ³n de los dÃ­as entresemana
# podrÃ­a deberse a que el fin de semana se descansa mÃ¡s.
```

## Coding Session II - 30 minutes

**Write a loop to plot 3 scatter plots of the following features:**

1. **`Minutes Lightly Active`  vs `Steps`**
1. **`Minutes Very Active` vs `Steps`**
1. **`Minutes Sedentary` vs `Steps`**

```python
# Driver 2 enter codes here
stepsdf = fitbit[['Minutes Lightly Active','Minutes Very Active','Minutes Sedentary','Steps']]
stepsdf
```

```python
fig,graphs = plt.subplots(figsize = (10,4), nrows = 1, ncols = 3)
for i in range(3):
    plt.sca(graphs[i])
    fig.tight_layout(pad=0.50) 
    plt.scatter(stepsdf.iloc[:,i],stepsdf['Steps'])
    plt.xlabel(stepsdf.iloc[:,i].name)
    plt.ylabel('Steps')
    plt.xticks(rotation=45);
    plt.grid(True)
```
